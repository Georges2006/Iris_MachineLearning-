```python
#Ce projet a pour objectif de construire un modèle de Machine Learning capable de classer 
# les espèces de fleurs d'iris (Setosa, Versicolor, Virginica). En utilisant le dataset Iris, 
# qui contient les mesures de la longueur et de la largeur des sépales et des pétales, nous allons entraîner un modèle Random Forest 
# Classifier. Ce modèle devra prédire correctement l'espèce d'une fleur d'iris en se basant sur ces quatre caractéristiques.

#Récupération des données
import pandas as pd
import numpy as np 
from matplotlib import  pyplot as plt 
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import label_binarize
feature = pd.read_csv('iris.csv' ,  sep=',')
print(feature)

#Néttoyage et exploration des données 
print(feature.info())  # vérifie qu'il y'a pas de valeur manquantes
print(feature.describe()) # donne une description statistique du dataset
print(feature.shape) # donne la taille du dataset 

#Modélisation
#séparation des données en données imput(X) et output(Y)
Y = np.array(feature['Species'])
X = feature.drop('Species',axis = 1 )
feature_list = list(X.columns)
X = np.array(X)
print(X.dtype)
#division en donnée de training et de test
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(X , Y, test_size = 0.3) #30% sont utilisé pour le test et 70% pour l'entrainement 

#Création du modèle 
from sklearn.ensemble import RandomForestClassifier 
clf = RandomForestClassifier(n_estimators=100) #nombre d'arbres
clf.fit(X_train , Y_train) #applique RFC aux données d'entrainement
Y_pred = clf.predict(X_test)
print("Accuracy" ,  metrics.accuracy_score(Y_test , Y_pred)) #Précison du modèle entre les résultats du test et de la  prédiction 
classes = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
Y_test_bin = label_binarize(Y_test , classes = classes)
Y_pred_bin =label_binarize(Y_pred , classes = classes)
auc = metrics.roc_auc_score(Y_test_bin, Y_pred_bin , average="macro")
print("AUC:", auc)










```
